```python
from os import walk
import os
# 打开文件

co_id_list = []
for root, dirs, files in os.walk("D:\pdf"):
    for name in dirs:
        print(os.path.join(root, name))
        co_id_list.append(name)
```

```python
zip_file = []
for root, dirs, files in os.walk("D:\pdf"):
    for name in files:
        if os.path.splitext(name)[1]=='.zip':
            zip_file.append(os.path.join(root,name))
for i in zip_file:
    print(i)
```

```python
other_file = []
for root, dirs, files in os.walk("D:\pdf"):
    for name in files:
        if os.path.splitext(name)[1]!='.pdf' and os.path.splitext(name)[1]!='.txt'  and os.path.splitext(name)[1]!='.doc' and os.path.splitext(name)[1]!='.zip':
            other_file.append(os.path.join(root,name))
```
