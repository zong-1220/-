```python
import glob

def getPath(fname):
    dirs = glob.glob('D:\pdf/*/'+fname+'/')
    if len(dirs) > 0:
        ret = dirs[0]
    else:
        ret = ""
    return ret

##測試

# spath = getPath('1101')
# if len(spath) > 0:
#     print("Found:"+spath)
# else:
#     print("No such path")
```

```python
from os import walk
import os
# 打开文件

co_id_list = []
for root, dirs, files in os.walk("D:\pdf下載"):
    for name in files:
        print(os.path.join(root, name))
        co_id_list.append(os.path.join(root, name))
```
